```python
from PIL import Image
import requests
import matplotlib.pyplot as plt
import numpy as np
from openpiv.piv import simple_piv
```

```python
def load_pair(url1=None, url2=None):
    # import images
    if url1 is None:
        url1 = 'http://raw.githubusercontent.com/OpenPIV/openpiv-python-examples/main/test6/piv01_1.bmp'
        url2 = 'http://raw.githubusercontent.com/OpenPIV/openpiv-python-examples/main/test6/piv01_2.bmp'
    
    response = requests.get(url1, stream=True)
    im1 = Image.open(response.raw)
    response = requests.get(url2, stream=True)
    im2 = Image.open(response.raw)
    return np.array(im1), np.array(im2)
```

```python
im1, im2 = load_pair()
```

```python
x,y,u,v,s2n = simple_piv(np.array(im1), np.array(im2), plot=True)
```

```python
from openpiv.tools import display_vector_field_from_arrays

display_vector_field_from_arrays(x,y,u,v,flags=s2n<1.4,mask=None);
                
```
